```python
import sys
from pathlib import Path
import json
```

```python
def test_efficientnet_keras(image_path, model_path, labels_path):
    print(f"\n--- Testing EfficientNet (.keras) ---")
    try:
        import numpy as np
        import tensorflow as tf
        from tensorflow.keras.preprocessing import image
        
        with open(labels_path, 'r', encoding='utf-8') as f:
            labels_data = json.load(f)
            labels = labels_data if isinstance(labels_data, list) else labels_data.get('labels', [])
            
        model = tf.keras.models.load_model(model_path)
        input_shape = model.input_shape
        h, w = input_shape[1:3]
        
        img = image.load_img(image_path, target_size=(h, w))
        x = image.img_to_array(img)
        x = np.expand_dims(x, axis=0)
        # EfficientNet in tf.keras often expects 0-255 inputs or preprocessed.
        # Assuming typical transfer learning pipeline.
        
        preds = model.predict(x)
        print(f"Raw predictions: {preds}")
        
        top_idx = np.argmax(preds[0])
        conf = preds[0][top_idx]
        label = labels[top_idx] if top_idx < len(labels) else f"class_{top_idx}"
        print(f"Predicted: {label} with confidence {conf:.2f}")
        
    except Exception as e:
        print(f"Error testing .keras model: {e}")
```

```python
image = "C:/Dev/projects/assist-eye/assets/model_testing/test_banknote.jpg"
keras = "C:/Dev/projects/assist-eye/assets/models/efficient_net/best_model_fine.keras"
labels = "C:/Dev/projects/assist-eye/assets/models/banknote_labels.json"
```

```python
test_efficientnet_keras(image, keras, labels)
```
