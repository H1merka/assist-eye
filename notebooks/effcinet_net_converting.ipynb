```python
import tensorflow as tf
```

```python
keras_model_path = 'C:/Dev/projects/assist-eye/assets/models/efficient_net/best_model_fine.keras'
tflite_model_path = 'C:/Dev/projects/assist-eye/assets/models/efficient_net/banknote_classifier_float16.tflite'
```

```python
model = tf.keras.models.load_model(keras_model_path)
converter = tf.lite.TFLiteConverter.from_keras_model(model)
```

```python
# Включаем оптимизации по умолчанию
converter.optimizations = [tf.lite.Optimize.DEFAULT]

# Указываем целевой тип float16 для весов
converter.target_spec.supported_types = [tf.float16]
```

```python
tflite_model = converter.convert()
```

```python
with open(tflite_model_path, 'wb') as f:
    f.write(tflite_model)
```
